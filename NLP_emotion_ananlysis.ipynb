# 数据处理部分



```python
#读取数据原始数据
import pandas as pd
filename="all_data.xlsx"
df=pd.read_excel(filename)
df.head()
```

```python
#删除重复值

df.drop_duplicates(inplace = True)

df
```

```python
#分词处理

import jieba

df['cut_jieba'] = df.content.apply(lambda x: ' '.join(jieba.cut(x)))

#抽取文摘的算法有TD-IDF， TextRank，后者的结果更为合理
df['abstract'] = df.content.apply(lambda x: ' '.join(jieba.analyse.textrank(x)))
df['TF-IDF_jieba'] = df.content.apply(lambda x: ' '.join(jieba.analyse.extract_tags(x)))


#关键词提取
df['Keywords'] = df.content.apply(lambda x: ' '.join(jieba.analyse.textrank(x, topK=5)))
df
```

```python
#数据预处理完毕，导出到excel文件

df.to_excel('20200222_data.xlsx', sheet_name = 'sheet1')
```
